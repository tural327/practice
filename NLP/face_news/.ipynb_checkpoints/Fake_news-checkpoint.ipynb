```python
import re
import os
import pandas as pd
import matplotlib.pyplot as plt 
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

## report results
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer

os.getcwd()
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head(3)
df = df.dropna()
df.reset_index(inplace=True)
```

```python
ps = PorterStemmer()
```

```python
X = []
text = df["text"]
len(text)
for i in range(0,len(text)):
    my_text = re.sub('[^a-zA-Z]', ' ', text[i])
    my_text = my_text.lower()
    my_text = my_text.split()
    
    my_text = [ps.stem(word) for word in my_text if not word in stopwords.words("english")]
    my_text = " ".join(my_text)
    X.append(my_text)
    ### check progress result
    work = (i*100)/len(text)
    
    print("Done {} %".format(round(work,2)))
```

```python
y=df['label']
```

```python
tfidf_v=TfidfVectorizer(max_features=5000,ngram_range=(1,3))
x_=tfidf_v.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train , X_test , y_train,y_test = train_test_split(x_,y ,test_size=0.2)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
matrix=confusion_matrix(y_test,y_pred)
print(matrix)
score=accuracy_score(y_test,y_pred)
print(score)
report=classification_report(y_test,y_pred)
print(report)
```
